```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Q1. You are a data analyst at an e-commerce company. Your task is to analyze sales data,
clean and process it, perform statistical analysis, and generate visual insights to help the
business make informed decisions.**

**Part I: Randomized Sales Data GeneraƟon (NumPy)**

```python
# 1. Initialize Random Seed
np.random.seed(102317039)

# 2. Generate Sales Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']

# Create a NumPy array with random integers between 1000 and 5000, shape (12, 4)
sales_data = np.random.randint(1000, 5001, size=(12, 4))

# 3. Convert to DataFrame
sales_df = pd.DataFrame(sales_data, columns=categories, index=months)

# Display the DataFrame
print(sales_df)
```

**Part II: Data Manipulation and Analysis (Pandas)**

```python
# 1. Display the first 5 rows and summary statistics
print(sales_df.head())
print(sales_df.describe())

```

```python
# 2. Calculate total sales per category and total sales per month
category_totals = sales_df.sum()
monthly_totals = sales_df.sum(axis=1)
print("Total sales per category:")
print(category_totals)
print("\nTotal sales per month:")
print(monthly_totals)
```

```python

# 3. Calculate average sales growth between consecutive months for each category
sales_growth = sales_df.pct_change().mean()
print("\nAverage sales growth per category:")
print(sales_growth)
```

```python
# 4. Add Total Sales and Growth Rate columns
sales_df['Total Sales'] = monthly_totals
sales_df['Growth Rate'] = sales_df['Total Sales'].pct_change() * 100
```

```python
# 5. Apply a 10% discount to Electronics if roll number is even
roll_number = int(input("Enter your roll number: "))
if roll_number % 2 == 0:
    sales_df['Electronics'] = sales_df['Electronics'] * 0.9  # 10% discount for even roll numbers
else:
    sales_df['Clothing'] = sales_df['Clothing'] * 0.85  # 15% discount for odd roll numbers
```

```python
# Display final DataFrame
print(sales_df)
```

**Part III: VisualizaƟons (Matplotlib and Seaborn)**

```python
plt.figure(figsize=(10, 5))
for category in categories:
    plt.plot(sales_df.index, sales_df[category], marker='o', label=category)
plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales Units")
plt.legend()
plt.grid()
plt.show()

# Box plots for sales distribution
plt.figure(figsize=(8, 5))
sns.boxplot(data=sales_df[categories])
plt.title("Sales Distribution by Category")
plt.xlabel("Product Categories")
plt.ylabel("Sales Units")
plt.show()
```

**Q.2 For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])**


```python
array = np.array([[1, -2, 3], [-4, 5, -6]])

# i. Element-wise absolute value
abs_array = np.abs(array)
print("\nElement-wise absolute values:")
print(abs_array)

# ii. Percentiles
percentiles_flat = np.percentile(array, [25, 50, 75])
percentiles_col = np.percentile(array, [25, 50, 75], axis=0)
percentiles_row = np.percentile(array, [25, 50, 75], axis=1)
print("\nPercentiles for flattened array (25th, 50th, 75th):", percentiles_flat)
print("\nPercentiles for each column:")
print(percentiles_col)
print("\nPercentiles for each row:")
print(percentiles_row)

# iii. Mean, Median, Standard Deviation
mean_flat = np.mean(array)
median_flat = np.median(array)
std_flat = np.std(array)

mean_col = np.mean(array, axis=0)
median_col = np.median(array, axis=0)
std_col = np.std(array, axis=0)

mean_row = np.mean(array, axis=1)
median_row = np.median(array, axis=1)
std_row = np.std(array, axis=1)

print("\nMean, Median, Std of flattened array:")
print(mean_flat, median_flat, std_flat)

print("\nMean, Median, Std for each column:")
print(mean_col, median_col, std_col)

print("\nMean, Median, Std for each row:")
print(mean_row, median_row, std_row)
```

**Q.3 For the array: a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0]).**

```python
import numpy as np

# Given array
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

# Compute floor, ceiling, truncated, and rounded values
floor_values = np.floor(a)       # Round down to nearest integer
ceiling_values = np.ceil(a)      # Round up to nearest integer
truncated_values = np.trunc(a)   # Remove decimal part (towards zero)
rounded_values = np.round(a)     # Round to the nearest integer

# Display results
print("Original Array:       ", a)
print("Floor Values:         ", floor_values)
print("Ceiling Values:       ", ceiling_values)
print("Truncated Values:     ", truncated_values)
print("Rounded Values:       ", rounded_values)
```

**Q.4 WAP to swap two elements in a list in Python Using a Temporary Variable.**

```python
# Function to swap elements using a temporary variable
def swap_elements(lst, index1, index2):
    temp = lst[index1]  # Store the first element in a temporary variable
    lst[index1] = lst[index2]  # Assign the second element to the first index
    lst[index2] = temp  # Assign the temporary variable to the second index

# Example usage
my_list = [10, 20, 30, 40, 50]
print("Original List:", my_list)

# Swap elements at index 1 and 3
swap_elements(my_list, 1, 3)

print("List after swapping:", my_list)
```

**Q.5 Write a Python program to swap two elements in a set by converƟng it to a list first.**

```python
my_set = {10, 20, 30, 40, 50}
print("Original Set:", my_set)

# Convert set to list
lst = list(my_set)

# Swap elements 20 and 40
i, j = lst.index(20), lst.index(40)
lst[i], lst[j] = lst[j], lst[i]

# Convert back to set
my_set = set(lst)

print("Set after swapping:", my_set)
```
