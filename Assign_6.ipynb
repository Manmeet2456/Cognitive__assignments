```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Question_1**

```python
M=int(input("Enter the value:"))
num_points = np.random.randint(5, 100)
x_values=np.linspace(-10,10,num_points)
y1=M*x_values**2
y2=M*np.sin(x_values)
plt.plot(x_values, y1, color='r', label='y = M*x^2') 
plt.plot(x_values, y2, color='b', label='y = M*sin(x)') 
plt.legend() 
plt.grid() 
plt.title('Mathematical Functions') 
plt.show()
```

**Question_2**

```python
A={
    "Subjects":["English","Hindi","Maths","Science","Socail Studies"],
    "Scores":[90,55,95,90,50]
}
result=pd.DataFrame(A)
ax=sns.barplot(result,x='Subjects',y='Scores',hue= 'Subjects' ,palette='Set1')
ax.set_title("Scores in Different Subjects", fontsize=14, fontweight='bold')
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')
plt.title("Subject Scores", fontsize=14, fontweight='bold')
plt.xlabel("Subjects", fontsize=12)
plt.ylabel("Scores", fontsize=12)
plt.xticks(rotation=30)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

**Question_3**

```python
roll_number = 102317039
np.random.seed(roll_number)

values = np.random.randn(50)

cumsum_values = np.cumsum(values)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(cumsum_values, marker='o', linestyle='-', color='b', label='Cumulative Sum')
axes[0, 0].set_title("Cumulative Sum Line Plot")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("Cumulative Sum")
axes[0, 0].legend()
axes[0, 0].grid(True, linestyle='--', alpha=0.7)

noise = np.random.randn(50) * 0.5
axes[0, 1].scatter(range(50), values + noise, color='r', alpha=0.6, label='Random Noise')
axes[0, 1].set_title("Scatter Plot with Random Noise")
axes[0, 1].set_xlabel("Index")
axes[0, 1].set_ylabel("Value")
axes[0, 1].legend()
axes[0, 1].grid(True, linestyle='--', alpha=0.7)

axes[1, 0].axis('off')
axes[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

**Question_4**

```python
data = pd.read_csv('company_sales_data.csv')

# 1. Line plot for Total Profit
data['month_number'] = data['month_number'].astype(str)  
plt.figure(figsize=(8, 5))
sns.lineplot(x='month_number', y='total_profit', data=data, marker='o', color='b', label='Total Profit')
plt.title("Total Profit Over Months")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.9)
plt.show()
```

```python
# 2. Multiline plot for product sales
plt.figure(figsize=(10, 6))
for column in data.columns[1:-1]:  # Skip 'month_number' and 'total_profit'
    sns.lineplot(x='month_number', y=column, data=data, marker='o', label=column)
plt.title("Product Sales Over Months")
plt.xlabel("Month Number")
plt.ylabel("Sales")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# 3. Bar chart for all attributes
plt.figure(figsize=(12, 6))
data.set_index('month_number').plot(kind='bar', figsize=(12, 6), colormap='viridis')
plt.title("Bar Chart for All Features")
plt.xlabel("Month Number")
plt.ylabel("Values")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
